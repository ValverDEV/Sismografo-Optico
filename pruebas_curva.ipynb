```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
x = np.linspace(0, 5, 100)
```

```python
def func(x, a, f):
    return a * np.sin((x/(2*f))*2*np.pi)
```

```python
a = 10
f = 2
y = func(x, a, f)
```

```python
plt.plot(x, y)
```

```python
amp, freq = curve_fit(func, xdata=x, ydata=y, p0 = [1, f])[0]
```

```python
y_fit = func(x, amp, freq)
```

```python
plt.plot(x, y)
plt.plot(x, y_fit, color = 'r')
```
